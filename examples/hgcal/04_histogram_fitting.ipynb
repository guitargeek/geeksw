```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.normal(size=10000)
```

```python
df = pd.DataFrame(data=dict(x=a))
```

```python
df.x.hist(bins=200)
plt.show()
```

```python
#bins = np.array([-4, -2, 0, 2, 4])
bins = np.linspace(-4, 4, 200)
```

```python
plt.hist(a, bins=bins)
plt.show()
```

```python
counts, bin_edges = np.histogram(a, bins=50)
```

```python
bin_centers = (bin_edges[1:] + bin_edges[:-1])/2.
```

```python
def f(x, mu, sigma, N):
    return N * np.exp(-(x - mu)**2/(2*sigma**2))
```

```python
from scipy.optimize import curve_fit
```

```python
popt, pcov = curve_fit(f, bin_centers, counts)
```

```python
plt.plot(bin_centers, f(bin_centers, *popt))
plt.hist(a, bins=50, histtype="step")
plt.show()
```

```python
q1, q2 = list(np.quantile(a, [0.16, 0.84]))
```

```python
plt.hist(a[np.logical_and(a > q1, a < q2)], bins=bins)
plt.show()
```
